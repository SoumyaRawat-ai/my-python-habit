```python
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'furnishingstatus', 'prefarea']
def binary_map(x):
    return x.map({'yes': 1, 'no': 0})

housing[varlist] = housing[varlist].apply(binary_map)
```

```python
housing.head()
```

```python
status = pd.get_dummies(housing['furnishingstatus']).astype(int)
status.head()
```

```python
status = pd.get_dummies(housing['furnishingstatus'], drop_first=True).astype(int)
```

```python
housing = pd.concat([housing, status], axis=1)
housing.head()
```

```python
housing.drop(['furnishingstatus'], axis=1, inplace=True)
housing.head()
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(housing, test_size=0.3, train_size=0.7, random_state=100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```
